```python
import csv
import pandas as pd
import plotly.express as px
import time


url_2019_01 = 'yellow_tripdata_2020-04.csv'


tip_amounts = []
payment_types = []


rownum = 0
with open(url_2019_01) as f:
  reader = csv.reader(f)
  for row in reader:
    if not (rownum == 0):
      try:
        if(int(row[9]) == 1):
          payment_types.append("Credit Card")
        elif(int(row[9]) == 2):
          payment_types.append("Cash")
        else:
          payment_types.append("Unknown / disputed ")
      except:
          payment_types.append("Unknown / disputed ")
      if(float(row[13]) < 0):
        tip_amounts.append(float(0))
      else:
        tip_amounts.append(float(row[13]))
    rownum += 1


tip_types_df = pd.DataFrame()
tip_types_df["Tip Amount"] = tip_amounts
tip_types_df["Payment Type"] = payment_types


tip_type_fig = px.box(tip_types_df, x="Payment Type", y="Tip Amount", notched=True)
tip_type_fig.show()
tip_type_fig.update_yaxes(range=[0,10])
```

```python
import plotly.io as pio


fh = open('tip_type_fig.html', 'w') # this just creates a blank file, so that the write_html() function can write to this file later
fh.close()

pio.write_html(tip_type_fig, file='tip_type_fig.html', auto_open=True, full_html=False) # preferred method
```
