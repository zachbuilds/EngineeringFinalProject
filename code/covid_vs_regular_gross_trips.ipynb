```python
import csv
import pandas as pd
import plotly.express as px
import time

url_2019_04 = 'apr_19.csv'
url_2020_04 = 'yellow_tripdata_2020-04.csv'

daily_trips_number = [0] * 60
years = [0] * 60
dates_in_april = [0] * 60


date = 1
for i in range(0,60,2):
  dates_in_april[i] = date
  dates_in_april[i+1] = date 
  date += 1


for i in range(0,60,2):
  years[i] = 2019
  years[i+1] = 2020


rownum = 0
with open(url_2019_04) as f1:
  reader = csv.reader(f1)
  for row in reader:
    if not (rownum == 0):
      if(int(row[1][5:7]) == 4):
        index = 2 * (int(row[1][8:10]) - 1)
        daily_trips_number[index] += 20
    rownum += 1


rownum = 0
with open(url_2020_04) as f1:
  reader = csv.reader(f1)
  for row in reader:
    if not (rownum == 0):
      if(int(row[1][5:7]) == 4):
        index = 2 * (int(row[1][8:10]) - 1) + 1
        daily_trips_number[index] += 1
    rownum += 1


covid_vs_regular_df = pd.DataFrame()
covid_vs_regular_df["Date in April"] = dates_in_april
covid_vs_regular_df["Total Number of Trips"] = daily_trips_number
covid_vs_regular_df["Year"] = years


gross_trips_fig = px.strip(covid_vs_regular_df, x="Date in April", y="Total Number of Trips", orientation="h", color="Year", title="Number of Trips for Each Day in April (Regular VS. COVID Year)")
gross_trips_fig.show()
```

```python
import plotly.io as pio


fh = open('gross_trips_fig.html', 'w') # this just creates a blank file, so that the write_html() function can write to this file later
fh.close()

pio.write_html(gross_trips_fig, file='gross_trips_fig.html', auto_open=True, full_html=False) # preferred method
```
