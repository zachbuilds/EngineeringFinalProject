```python
import csv
import pandas as pd
import plotly.express as px
import time


def getDayNum(dayname):
  days_dict = {"Sunday":0, "Monday":1, "Tuesday":2, "Wednesday":3, "Thursday":4, "Friday":5, "Saturday":6}
  return days_dict.get(dayname)

sunday_hourly_trips = [0] * 24
monday_hourly_trips = [0] * 24
tuesday_hourly_trips = [0] * 24
wednesday_hourly_trips = [0] * 24
thursday_hourly_trips = [0] * 24
friday_hourly_trips = [0] * 24
saturday_hourly_trips = [0] * 24

collected_trip_data = [sunday_hourly_trips, monday_hourly_trips, tuesday_hourly_trips, wednesday_hourly_trips, thursday_hourly_trips, friday_hourly_trips, saturday_hourly_trips]


url_2019_01 = 'yellow_tripdata_2020-04.csv'


rownum = 0
with open(url_2019_01) as f:
  reader = csv.reader(f)
  for row in reader:

    if not (rownum == 0):

      current_timestamp = row[1]
      current_timestamp_date = current_timestamp[:10]
      current_timestamp_time = current_timestamp[11:]


      temp_timestamp_date_obj = pd.Timestamp(current_timestamp_date)
      current_dayname = temp_timestamp_date_obj.day_name()
      current_daynum = getDayNum(current_dayname)


      current_hour_str = current_timestamp_time[0:2]
      current_hour = int(current_hour_str)


      collected_trip_data[current_daynum][current_hour] += 1


    rownum += 1


popular_ridetimes_fig = px.imshow(collected_trip_data,
                labels=dict(x="Day of Week", y="Time of Day", color="Number of Trips"),
                y=['Sunday', 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday'],
                x=["12 am", "1 am", "2 am", "3 am", "4 am", "5 am", "6 am", "7 am", "8 am", "9 am", "10 am", "11 am", "12 pm", "1 pm", "2 pm", "3 pm", "4 pm", "5 pm", "6 pm", "7 pm", "8 pm", "9 pm", "10 pm", "11 pm"],
                title = "Aggregate Day/Time Data For the Month of April in 2020"
               )
popular_ridetimes_fig.update_xaxes(side="bottom")
popular_ridetimes_fig.show()
```

```python
import plotly.io as pio


fh = open('popular_ridetimes_fig.html', 'w') # this just creates a blank file, so that the write_html() function can write to this file later
fh.close()

pio.write_html(popular_ridetimes_fig, file='popular_ridetimes_fig.html', auto_open=True, full_html=False) # preferred method
```
