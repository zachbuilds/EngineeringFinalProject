```python
import csv
import pandas as pd
import plotly.express as px
import time


url_2019_04 = 'apr_19.csv'
url_2020_04 = 'yellow_tripdata_2020-04.csv'


total_number_trips_2019_04 = 0
total_number_trips_2020_04 = 0
total_tippers_2019_04 = 0
total_tippers_2020_04 = 0


rownum = 0
with open(url_2019_04) as f1:
  reader = csv.reader(f1)
  for row in reader:
    if not (rownum == 0):
      if(float(row[7]) > 0):
        total_tippers_2019_04 += 1
      total_number_trips_2019_04 += 1
    rownum += 1

total_tippers_2019_04 *= 20
total_trips_2019_04 *= 20

rownum = 0
with open(url_2020_04) as f2:
  reader = csv.reader(f2)
  for row in reader:
    if not (rownum == 0):
      if(float(row[13]) > 0 ):
        total_tippers_2020_04 += 1
      total_number_trips_2020_04 += 1
    rownum += 1


data_2019 = dict(
    number=[total_number_trips_2019_04, total_tippers_2019_04],
    group=["Passengers", "Tippers"])
tipper_proportion_2019_fig = px.funnel(data_2019, x='number', y='group', title="April 2019 (Pre-COVID)")
tipper_proportion_2019_fig.show()


data_2020 = dict(
    number=[total_number_trips_2020_04, total_tippers_2020_04],
    group=["Passengers", "Tippers"])
tipper_proportion_2020_fig = px.funnel(data_2020, x='number', y='group', title="April 2020 (Beginning of COVID)")
tipper_proportion_2020_fig.show()
```

```python
import plotly.io as pio


fh = open('tipper_proportion_2020_fig.html', 'w') # this just creates a blank file, so that the write_html() function can write to this file later
fh.close()

pio.write_html(tipper_proportion_2020_fig, file='tipper_proportion_2020_fig.html', auto_open=True, full_html=False) # preferred method
```
