```python
import csv
import pandas as pd
import plotly.express as px
import time


url_2019_01 = 'jan_19.csv'


taxi_zone_numbers = [n for n in range(1,266)]
pickup_zone_frequencies = [0] * 265
dropoff_zone_frequencies = [0] * 265


rownum = 0
with open(url_2019_01) as f:
  reader = csv.reader(f)
  for row in reader:
    if not (rownum == 0):
      pickup_zone_frequencies[int(row[7]) - 1] = pickup_zone_frequencies[int(row[7]) - 1] + 1
      dropoff_zone_frequencies[int(row[8]) - 1] += 1
    rownum += 1
 

df = pd.DataFrame()
df["Zone"] = taxi_zone_numbers
df["Pickup Count"] = pickup_zone_frequencies
df["Dropoff Count"] = dropoff_zone_frequencies


pickup_fig = px.bar(df, x='Zone', y='Pickup Count', title="Pickup Counts For Each Zone in January of 2019")
pickup_fig.update_layout(xaxis = dict(tickmode = 'array', tickvals = taxi_zone_numbers))
pickup_fig.update_xaxes(tickangle=-70, tickfont_size=4, automargin=True)
pickup_fig.show()


dropoff_fig = px.bar(df, x='Zone', y='Dropoff Count', title="Dropoff Counts For Each Zone in January of 2019")
dropoff_fig.update_layout(xaxis = dict(tickmode = 'array', tickvals = taxi_zone_numbers))
dropoff_fig.update_xaxes(tickangle=-70, tickfont_size=4, automargin=True)
dropoff_fig.show()
```

```python
import plotly.io as pio


fh = open('dropoff_fig.html', 'w') # this just creates a blank file, so that the write_html() function can write to this file later
fh.close()

pio.write_html(dropoff_fig, file='dropoff_fig.html', auto_open=True, full_html=False) # preferred method
```
