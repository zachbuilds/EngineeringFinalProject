```python
import csv
import pandas as pd
import plotly.express as px
import time

url_2020_04 = 'yellow_tripdata_2020-04.csv'


trip_distances = []
tip_amounts = []
passenger_counts = []
months = []


rownum = 0 
with open(url_2020_04) as f:
  reader = csv.reader(f)
  for row in reader:
    if(rownum != 0):
      if(rownum % 100 == 0 and row[3]):
        trip_distances.append(float(row[4]))
        tip_amounts.append(float(row[13]))
        passenger_counts.append(float(row[3]))
        months.append(4)
    rownum += 1


duration_tip_df = pd.DataFrame()
duration_tip_df["Trip Distance"] = trip_distances
duration_tip_df["Tip Amount"] = tip_amounts
duration_tip_df["Month"] = months
duration_tip_df["Passenger Count"] = passenger_counts


duration_passenger_tip_fig = px.parallel_coordinates(duration_tip_df,
                              dimensions=['Trip Distance', 'Tip Amount', 'Passenger Count'])
duration_passenger_tip_fig.show()
```

```python
import plotly.io as pio


fh = open('duration_passenger_tip_fig.html', 'w') # this just creates a blank file, so that the write_html() function can write to this file later
fh.close()

pio.write_html(duration_passenger_tip_fig, file='duration_passenger_tip_fig.html', auto_open=True, full_html=False) # preferred method
```
